K-Means Using a Refinement Algorithm
=====================================

Though K-Means is a simple and powerful algorithm for searching for clusters, it suffers from two significant drawbacks:

1. The problem of finding the global minimum for K-Means is NP-hard,
2. Finding a local solution is highly sensitive to the initial vector specification, with respect to both speed of convergence and minimizing distortion. 

Consequently, the goal of the refinement algorithm is to begin by finding a better place to start the search, by taking small subsamples of the data and running a modified K-Means search on each subsample. The results of these operations are combined into a new subset, and K-Means is then applied to this subset, with each set of centroids from the original subsample as a starting point. The K-Means iteration which produces the least distortion is then chosen as the refined initial conditions to use for K-Means on the original data. 

This refinement algorithm achieves several things:
- **It reduces sensitivity to a random starting point**: Though each individual K-Means is instantiated from a random point over the sub-samples, the refinement process smoothes over differences caused by starting at outliers in the data. K-Means applied to the same data multiple times with refinement is inherently more stable than K-Means applied without, with respect to both distortion and number of iteratons. 
- **It reduces the amount of iterations of K-Means**: Because K-Means usually is initialized from random points in the data, there is no control over where you start out. Thus, a bad starting location might require more iterations of K-Means to converge. Over small data sets this is not a cause of too much concern. However, over large and especially distributed data stores, a bad initialization can be expensive, especially cosidering that the complexity of running K-Means increases in polynomial time as the amount of observations increase. By choosing a starting point from K-Means applied to smaller subsets of the data, one can guess at a starting point that is closer to the real centroids, and therefore reduce the number of iterations needed to converge. 
- **It reduces the amount of distortion**: In their paper, the authors find that algorithm reduces distortion between the K-Means result and the centroids. Given that K-Means is deterministic once a starting point has been specified, this makes sense: if your initial guess is more accurate, then you're already closer to finding your solution, and presumably closer to being accurate. 


As the article points out, most common initialization methods (such as a random partition start) do no better than a Forgy random start. Although the authors don't lay out a formal proof for the reduced time complexity of their algorithm, empirically, they show how their algorithm is faster than classic K-Means over large data sets. 

Intuition demonstrates how this algorithm is certainly superior to classic K-Means. By way of illustration, suppose a friend of yours invites you and your friends to a party at a large club. When you arrive, the party host hands you a handkerchief and informs you that at this party, everyone has to wear a blindfold to be admitted. You want to be in the middle of your friend group, but the only way you for you to find your friends is to ask each and every person their name until you find your friend group. This of course would take a lot of time and be really embarassing. The host says that she will take you to a random spot in the hall and leave you to wander from there. Alternatively, she can offer to give you a random list of names from the guestbook, and she will take you to the person you choose from the list. Which option should you choose? 

Clearly its best to have some information about where to start than no information. It would reduce the amount of people you would have to query for names, and hopefully land you closer to your friends. Roughly, the refinment algorithm is like the host who offers you a random sample of names. The additional information cuts down on the total work needed to arrive at a solution. 

Finally, the paper gives the broad strokes of what a proof might look like: Let $D$ be the set of observations in the data and $S$ be a small subsample, and let $iter(D)$ be the number of iterations of K-Means needed to converge to a solution in $D$. Since $|S| << |D|$, we can conjecture that $iter(S) << iter(D)$, and that the byproduct of the refinement algorithm produces an initial vector $v_0$ which reduces $iter(D)$, because we presume that $v_0$ is an approximation to the true centroids. Thus, the time complexity of the refined algorithm is less than the time complexity of the classic algorithm, with a certain probability. 

## Code for the K-Means Refinement Algorithm
Below is a pseudo-code sketch of the refinement algorithm, and immediately following is a python implementation. Note that this algorithm assumes that you can take truly random subsamples from a data store. This is not a trivial problem considering how data can be stored, indexed and ordered in a variety of ways. 
In addition, this algorithm assumes a K-Means implementation, a sampling function and a distortion function.

### In Pseudo-Code

```pseudocode

// Objective: 
// Find an initial starting vector of size K to seed a potentially expensive k-means search by taking j random subsamples of 
// the data and applying a refinement algorithm

begin refinement

assume data
assume a function k-means which computes the classic k-means algorithm over a given data set with random start
assume a function random_sample(data, ratio) which takes data and returns a % of the sample with replacement
assume a function distortion 
            which computes the sum of squared euclidean distances between a point in the data and its closest cluster
            
def kmeans-mod(data, starting_point, k-clusters):
    let output be the result of classic k-means applied to data with k-clusters at starting point
    let new_centroids be a new vector of starting points
    if in output there exists a cluster without any members:
        for each cluster k:
            if k has no members:
                add add the point farthest from its cluster centroid to new_centroids
            else:
                add k to new_centroids
        return kmeans(data, new_centroids, k-clusters)
    else:
        return output

let k = the number of clusters to fit
let j = the number of random subsamples to take
let r = the size of the subsamples as a percentage of the total
let J be a set of size j of random samples
let CM be an empty set  // name as referenced in the paper
let FM be an empty set  // name as referenced in the paper

do j times
    random_sample(data, r) and add to J

for each random sample in J:
    compute k-means-mod with random start and k clusters
    add result to CM
    
for each result of CM:
    compute k-means over all CM with result as starting vector and k clusters
    add result to FM
    
return the centroids of a solution in FM such that distortion(solution) is minimized
    
end refinement
```

### In Python
The following is my own implementation of the algorithm in python. The only dependency is numpy. While scipy does have two implementations of k-means to choose from, neither of the two return enough information to implement k-means-mod. As a result, I had to implement k-means from the ground up. My implementation is certainly not as fast as the scipy implementation. That being said, the refinement algorithm is still demonstrably faster and more accurate than classic k-means, as shown by the graphs included below.


````python
class Point(object):
    def __init__(self, vec):
        self._data = vec
        self._cluster = None

    def __getitem__(self, value):
        return self._data[value]
    
    def __repr__(self): return np.array_str(self._data)
    
    def distance_from(self, point):
        return euclidean(self._data, point._data)**2 # Squared euclidean distance
    
    def closest_centroid(self, centroids):
        """Given a CentroidSet, returns the closest centroid."""
        key_val = ((k, self.distance_from(v)) for k, v in centroids)
        
        def min_idx(a,b):
            """Assumes first entry of tuple is index and second is value"""
            if a[1] < b[1]:
                return a
            elif a[1] > b[1]:
                return b
            else:
                return a
                
        return reduce(min_idx, key_val)
    
    def assign_cluster(self, centroids):
        """Takes a CentroidSet and assigns a cluster to the object inplace. Returns nothing."""
        self._cluster = self.closest_centroid(centroids)[0]

    
class DataPoint(Point):
    
    @property
    def cluster(self):
        return self._cluster
    
    @cluster.setter
    def set_cluster(self, value):
        self._cluster = value
    

class Centroid(Point): 
    pass


class DataSet(object):
    def __init__(self, mat):
        self._points = {idx: DataPoint(row) for (idx, row) in enumerate(mat)}
        
    def __iter__(self):
        return self._points.itervalues()
    
    def __getitem__(self, idx):
        return self._points[idx]
    

class CentroidSet(object):
    def __init__(self, k, dim, vec0=None):
        self._centroids = {idx: Centroid(row) for idx, row in enumerate(vec0)} if vec0 is not None else {}
        self._k = k
        self._dim = dim
    
    def __iter__(self):
        return self._centroids.iteritems()
    
    def __eq__(self, other):
        testeq = lambda i: np.array_equal(self._centroids[i]._data, other._centroids[i]._data)
        return all(testeq(i) for i in range(self._k))
    
    def __ne__(self, other):
        return not self.__eq__(other)
    
    def random_start(self):
        for i in range(self._k):
            self._centroids[i] = Centroid(np.random.poisson(7, self._dim))
    
    def clusters(self, data):
        out_dict = {idx:[] for idx in range(self._k)}
        for pt in data:
            out_dict[pt.cluster].append(pt)
        return out_dict
        
    def generate_centroids(self, data):
        new_cents = {}
        c = self.clusters(data)
        for idx, ls in c.iteritems():
            new_vec = np.mean([i._data for i in ls], axis=0)
            new_cents[idx] = Centroid(new_vec) if not np.isnan(np.sum(new_vec)) else self._centroids[idx]
        return new_cents
    
    def recompute_centroids(self, data):
        self._centroids = self.generate_centroids(data)
        
        
def my_kmeans(data, start_points, k):
    """Takes a numpy array as data, a numpy array of starting centroids and number of clusters K 
    and returns a K-Mean solution tuple where the first entry is a dictionary of the original data separated 
    into K clusters, and the second entry are the cluster centroids."""
    centroids = CentroidSet(k, data.shape[1], vec0=start_points)
    data      = DataSet(data)
    new_vecs  = centroids
    while True:
        old_vecs = deepcopy(centroids)
        for pt in data:
            pt.assign_cluster(new_vecs)
        new_vecs.recompute_centroids(data)
        if old_vecs == new_vecs:
            return new_vecs.clusters(data), new_vecs._centroids

def get_empty_centroids(k_means_tupl):
    """Takes the output tuple of my_kmeans and returns a dictionary of clusters that are emptpy, 
    along with the centroid of the empty cluster."""
    empties = {}
    clustered_data, centroids = k_means_tupl
    for key, val in clustered_data.iteritems():
        if not val:
            empties[key] = centroids[key]
    return empties

def results_to_pair(results):
    """Converts k-means results into a list of 2-tuples where the first item is the point and the second is its cluster centroid."""
    clusters, centroids = results
    out_list = []
    for i in centroids:
        for cluster in clusters[i]:
            out_list.append((cluster, centroids[i]))
    return out_list

def points_sorted_by_distance_from_center(results):
    return [pair[0] for pair in sorted(results_to_pair(results), 
                                       key=lambda (a,b): a.distance_from(b))]

def dict_to_np_array(dict_of_points):
    """Converts a dictionary of numpy arrays into a numpy matrix array."""
    return np.array([dict_of_points[i]._data for i in range(len(dict_of_points))])

def my_kmeans_mod(data, start_points, k):
    """A modified version of K-Means, as described in the article."""
    results = my_kmeans(data, start_points, k)
    
    if get_empty_centroids(results):
        
        new_init_dict   = {}         # Create a new dictionary that will hold our initial conditions
        all_centroids   = results[1] # Get the dict of all the centroids from the resulsts of vanilla kmeans
        empty_centroids = get_empty_centroids(results) # Get the dict of empty centroids from results
        new_points      = points_sorted_by_distance_from_center(results)
        
        for key in all_centroids:
            if empty_centroids.get(key):
                new_init_dict[key] = new_points.pop()._data
            else:
                new_init_dict[key] = all_centroids[key]._data
                
        # convert new_init_dict to numpy vector that we can pass to k_means()
        new_start = np.array([new_init_dict[i] for i in range(len(new_init_dict))]) # Convoluted, but done to preserve the order
        return my_kmeans(data, new_start, k)
    else:
        return results
 
def distortion(k_means_results):
    """Takes the output tuple of K-Means and returns a messure of distortion."""
    data_pts, centroids = k_means_results
    accum = 0
    for key in range(len(centroids)):
        accum += sum(euclidean(centroids[key]._data, i._data)**2 for i in data_pts[key])
    return accum

def rand_sample(data, ratio):
    if ratio < 1:
        obs_nums = int(len(data)*ratio)
    else:
        obs_nums = ratio
    idx = np.arange(len(data))
    np.random.shuffle(idx)
    return data[idx[:obs_nums]]

def j_rand_sample(data, ratio, j):
    """Takes j random samples from data at the specified ratio. Returns a list of numpy arrays."""
    return [rand_sample(data, ratio) for i in range(j)]

def refinement_alg(data, k_clusters, j_subsets):
    constant_k_means = lambda data: my_kmeans_mod(data, rand_sample(data, k_clusters), k_clusters)
    J = map(constant_k_means, j_rand_sample(data, 0.01, j_subsets))
    CM = np.array([i for K in J for i in dict_to_np_array(K[1])])
    candidates = [my_kmeans(CM, dict_to_np_array(CM_i[1]), k_clusters) for CM_i in J]
    return min(candidates, key=distortion)[1]
    
````

````python
from numpy.random import seed
seed(1337)

# Make data from 3 gaussian distributions
data = np.concatenate([np.random.normal((20, 5), 2, (300,2)), 
                       np.random.normal((0 ,10), 2, (300,2)), 
                       np.random.normal((15,15), 2, (300,2))], axis=0)

# Compute both algorithms
K = 3    # The number of clusters
J = 10   # The number of random subsamples
res_a = my_kmeans(data, dict_to_np_array(refinement_alg(data, K, J)), K)
res_b = my_kmeans(data, rand_sample(data, K), K)
print "Refined distortion: ", distortion(res_a)
print "Non-Refined distortion: ", distortion(res_b)
````

### Results of running K-Means with refinement (red) and without refinement (green).

````python
res0 = res_a[1]
res1 = res_b[1]

# Red is refined k-means, green is vanilla k-means
sns.plt.figure(figsize=(10,6))
sns.plt.scatter(data.T[0], data.T[1], color='lightgray')
for i in range(3):
    sns.plt.scatter(*res0[i], color='red')
    sns.plt.scatter(*res1[i], color='green')
````
