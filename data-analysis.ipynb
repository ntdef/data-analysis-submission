```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from datetime import datetime

# Helper function for convertng MeFi's dates into Pandas compatible datetimes
def convert_datetime(date_as_string):
    fmt_string = '%b %d %Y %I:%M:%S:%f%p'
    return datetime.strptime(date_as_string, fmt_string)
```

```python
# Grab the data.
tag_data = pd.read_table('tagdata_askme.txt', skiprows=1) # Skip the first row; contains meta-data  
post_data = pd.read_table('postdata_askme.txt', skiprows=1)
```

### Clean the Post Data

```python
# Check for missing data...
post_data.loc[post_data.isnull().any(1).nonzero()[0],:]
```

```python
# Looks like [NULL] is the infodump's special code for NA.
# fill any blanks inreason col with [NULL]
post_data.reason = post_data.reason.fillna('[NULL]') 
```

```python
# Recheck for missing data
post_data.loc[post_data.isnull().any(1).nonzero()[0],:]
```

```python
# Looks like they are all null categories posted by one used id. 
# We're going to drop those
post_data.dropna(inplace=True)
```

```python
# Categories should be integers, not floats
post_data.category = post_data.category.map(int)
```

```python
# Convert dates to python datetime
post_data.datestamp = post_data.datestamp.map(convert_datetime)
```

```python
# These labels are taken from the MetaFilter InfoDump wiki page
categories = [(1, 'computers & internet', 'computers-internet'),
 (2, 'technology', 'technology'),
 (3, 'home & garden', 'home-garden'),
 (4, 'work & money', 'work-money'),
 (5, 'sports, hobbies, & recreation', 'sports-hobbies-recreation'),
 (6, 'society & culture', 'society-culture'),
 (7, 'travel & transportation', 'travel-transportation'),
 (8, 'science & nature', 'science-nature'),
 (9, 'education', 'education'),
 (10, 'health & fitness', 'health'),
 (11, 'shopping', 'shopping'),
 (12, 'food & drink', 'food-drink'),
 (13, 'writing & language', 'writing-language'),
 (14, 'human relations', 'human-relations'),
 (15, 'media & arts', 'media-arts'),
 (16, 'pets & animals', 'pets-animals'),
 (17, 'religion & philosophy', 'religion-philosophy'),
 (18, 'clothing, beauty, & fashion', 'clothing-beauty-fashion'),
 (19, 'law & government', 'law-government'),
 (20, 'grab bag', 'grab-bag')
 ]

labels = pd.DataFrame(categories, columns=['category', 'label', 'url_tag'])
labels.set_index('category', inplace=True)
labels.drop('url_tag', 1, inplace=True)
```

```python
post_data = post_data.set_index('category').join(labels).reset_index(drop=True)
post_data.columns = list(post_data.columns[:-1]) + ['category']
```

```python
# There are 21 different categories
post_data.category.value_counts(dropna=False)
```

### Clean the Tag Data

```python
# Clean the tag data
idx = tag_data.isnull().any(1).nonzero()[0]
tag_data.ix[idx]
```

```python
# Only NAs are tags without names, so we'll drop those
tag_data.dropna(inplace=True)
```

```python
# Convert to python datetime
tag_data.link_date = tag_data.link_date.map(convert_datetime)
```

```python
# Mutltiple posts at the same exact time.
(post_data.datestamp.value_counts() > 1).value_counts()

# Because tags are joined by exact data, we have to throw out posts that have colliding post times 
```

```python
# Throw out entries where there are colliding datestamps
print "The number of entries I'm discarding is: ", len(post_data.groupby('datestamp').filter(lambda x: len(x) > 1))
post_data = post_data.groupby('datestamp').filter(lambda x: len(x) < 2)
```

### What were the top 3 deleted categories?

The top three deleted categories were __human relations__, __computers & internet__ and __society & culture__.

```python
counts_deleted = post_data.groupby('category')['deleted'].sum()
```

```python
# Looks like category 14 was most likely to be deleted
counts_deleted.sort(ascending=False)
counts_deleted.plot(kind='bar', figsize=(10,6), color=['#e74c3c']*3 + ['#34495e']*17)
sns.axlabel('category', 'number of posts deleted')
```

```python
# now as a ratio of total
counts_deleted/sum(counts_deleted)
# Category 14 was deleted 20% of the time
```

### What were the top 3 most commented categories?

The top three most commented categories were __human relations__, __computers & internet__ and __media & arts__

```python
comment_counts = post_data.groupby('category')['comments'].sum()
comment_counts.sort(ascending=False)
comment_counts.plot(kind='bar', 
                    figsize=(10,6),
                    color=['#e74c3c']*3 + ['#34495e']*17)
sns.axlabel('category', 'number of comments')
```

### What were the top 3 most favorited categories?

The top three favorited categories were __media & arts__, __human relations__ and __food & drink__.

```python
fav_counts = post_data.groupby('category')['favorites'].sum()
fav_counts.sort(ascending=False)
fav_counts.plot(kind='bar', figsize=(10,6),
                    color=['#e74c3c']*3 + ['#34495e']*17)
sns.axlabel('category', 'number of favorites')
```

### Which tags were most popular overall?

By far the most popular tag was __resolved__

```python
# Top twenty
tag_counts = tag_data.tag_name.value_counts()
tag_counts[:20].plot(kind='bar', figsize=(10,6))
sns.axlabel('tag', 'number of times applied')
```

### The top tag only accounts for 3.8% of the total tags applied

```python
# Top twenty, as ratio of total counts
(tag_counts/tag_counts.sum())[:20]
```

### Joining to post data

```python
clean = pd.merge(post_data, 
                 tag_data, 
                     left_on=['datestamp', 'postid'], 
                     right_on=['link_date', 'link_id'],
                     how='left')

clean = clean[['postid', 'userid', 'datestamp', 'category', 'comments', 'favorites', 'deleted', 'tag_name']]
```

### What are the most favorited tags?

However, __music__ was the most favorited

```python
tag_favs = clean.groupby('tag_name')['favorites'].sum()
tag_favs.sort(ascending=False)
tag_favs[:20].plot(kind='bar', figsize=(10,6))
sns.axlabel('number of times favorited', 'tag')
```

### What tags garnered the most comments?

```python
tag_comments = clean.groupby('tag_name')['comments'].sum()
tag_comments.sort(ascending=False)
tag_comments[:20].plot(kind='bar', figsize=(10,6))
sns.axlabel('tag', 'number of comments')
```

## How would you characterize the distribution of favorites?

The distribution of favorites is strongly unimodal, with significant tails. The mean number of favorites was 5 with a standard deviation of 15, indicating that the distribution is very sparse.
Nearly 99% of posts had less than 50 favorites, and 34% had exactly 0. There were 3 outliers that had more than 1000 favorites, shown below.

```python
post_data.favorites.describe()
```

```python
# What percentage had less than 50 favs?
len(post_data.favorites[post_data.favorites < 50])/float(len(post_data.favorites))
```

```python
# What percentage had exactly 0 favs?
len(post_data.favorites[post_data.favorites==0])/float(len(post_data.favorites))
```

```python
# Outliers with more than 1000 favorites
post_data[post_data.favorites > 1000]
```

## How would you characterize the distribution of comments?

The distribution of comments is similar to the distribution of favorites: unimodal with a long tail. The average post had 14 comments, with a standar deviation of 14. Among all the posts, 75% had less than 18 comments. However, 99% posts had at least one comment, and only 2 posts had more than 500 comments.

```python
post_data.comments.describe()
```

```python
# What percentage had more than comments?
(post_data.comments > 500).value_counts(True)
```

The following kernel desntiy estimate plot shows an estimate of the probability density function for the distribution of comments. As you can see from the graph, most of the posts have less than 100 comments.

```python
sns.kdeplot(post_data.comments, shade=True)
```

## Distribution of Comments & Favorites with respect to the Categories

```python
# Counts of comments and favorites by category
totals = post_data.groupby('category')['comments', 'favorites'].sum()
totals
```

Below is a bar graph showing the share of total comments and favorites for each cateogry. For instance 15% of all comments occured on posts categorized as __human relations__. __Media & arts__ and __food & drink__ were favorited siginificantly more than they were commented.

```python
totals.apply(lambda col: col.div(col.sum(axis=0))).sort('comments', ascending=False).plot(kind='bar', figsize=(12,6))
```

Below is a distribution graph of comments by category for posts with less than 70 comments. On the x axis are the counts of comments per post and on the y axis is the probability of observing the x comments. This chart shows that the categories are all roughly unimodally distributed, althoughh the variance varies from category to category. For instance, __human relations__ has a wide spread of comments, whereas __computers & internet__ posts are clustered around ~5 comments.

```python
sub_data = post_data[(post_data["comments"] < 70)]
g = sns.FacetGrid(sub_data, row='category', size=1.7, aspect=4)
g.map(sns.distplot, "comments")
```

## Bonus: 
Did __work & money__ posts grow as a share of total posts on MetaAsk after the 2008 Recession?

```python
total = clean.set_index('datestamp').resample('QS', how='count')['postid']
cat4 = clean[clean['category'] == "work & money"].set_index('datestamp').resample('QS', how='count')['postid']
```

```python
sns.plt.figure(figsize=(16,6))
(cat4/total).plot()
sns.plt.axvline('2008-12', ymax=0.9, color='gray', linestyle='--')
sns.plt.text('2008-12', 0.104, '2008 Recession \nStarts', ha='center')
sns.axlabel('post date by quarter', 'share as a ratio of all posts')
```
